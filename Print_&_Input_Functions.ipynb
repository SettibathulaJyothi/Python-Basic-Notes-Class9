# **The Print() Function**
* The print() function is used to printan output on the screen
* It converts the expression into a string before writing to the screen
* Syntax:print(object(s),sep=separator,end=end)
  * where *object* can be more or one seperated by comma and it can be a variable.Object is converted to string before printing
  * where *sep* seperator between the items to print,Default sep is space
  * where *end* allows to specify what to print at the end of the output.default is newline character ('\n') means each print statement ends with a nwe line

**Following are the examples of the print() function**


```python
print("hello")
```

```python
print("hello","friends")
```

```python
print("hello"+"friends")#(This is concatenation of string using "+" operator)
```

```python
print("hello","friends",sep="$")
```

```python
print("hello","friends",sep="$",end="!")
```
